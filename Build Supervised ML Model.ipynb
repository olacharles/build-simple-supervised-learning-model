### bar chart function

```python
def dsc_chart(feature):
    absent = train[train['Absent']==1][feature].value_counts()
    notabsent = train[train['Absent']==0][feature].value_counts()
    df = pd.DataFrame([absent, notabsent])
    df.index=['Absent', 'Present']
    df.plot(kind='bar', stacked=True, figsize=(10,8))
    plt.ylabel(feature)
    plt.xlabel('Attendance Status')
```

# Feature Engineering

```python
def Care(train):
    if train.Pet >=1 and train.Son >=1:
        return "High"
    elif train.Pet<1 and train.Son >=1:
        return "Moderate"
    elif train.Pet >= 1 and train.Son <1:
        return 'Moderate'
    else:
        return 'No care'
train['CareDemand'] = train.apply(lambda train:Care(train), axis=1)

def Care(test):
    if test.Pet >=1 and test.Son >=1:
        return "High"
    elif test.Pet<1 and test.Son >=1:
        return "Moderate"
    elif test.Pet >= 1 and test.Son <1:
        return 'Moderate'
    else:
        return 'No care'
test['CareDemand'] = test.apply(lambda test:Care(test), axis=1)
```

## Density Plot

```python
def kdplot(feature):
    facet = sns.FacetGrid(train, hue = 'Absent', aspect = 5)
    facet.map(sns.kdeplot, feature, shade=True)
    facet.set(xlim=(train[feature].min(), train[feature].max()))
    facet.add_legend()
```

```python
kdplot('Transportation expense')
kdplot('Distance from Residence to Work')
kdplot('Age')
```

# Modelling

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

binary = ['Dist', 'Disciplinary failure', 'Social drinker', 'Social smoker']
multi = ['Month of absence', 'Day of the week', 'Age_group', 'TransportFare', 'Education', 'CareDemand']

le = LabelEncoder()
for i in binary:
    train[i] = le.fit_transform(train[i])
for i in binary:
    test[i] = le.fit_transform(test[i])
    
train = pd.get_dummies(data= train, columns = multi)
test = pd.get_dummies(data = test, columns = multi)
```

```python
features = train.drop('Absent', axis=True)
target = train['Absent']
```

## Scaling Features and test data

```python
scaler = StandardScaler()
scaler.fit(features)
scaler.fit(test)
```

## Split Train and Test data

 This test data here is different from the test data to be submitted on Kaggle

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30, random_state = 30)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
print('Train data score: ', clf.score(X_train,y_train))
print('Validation data score: ', clf.score(X_test,y_test))
```

## Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression()
clf2.fit(X_train,y_train)
print('Train data score: ', clf2.score(X_train,y_train))
print('Validation data score: ', clf2.score(X_test,y_test))
```
